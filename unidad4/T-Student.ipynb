```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

data = np.array([118, 125, 130, 128, 122, 135, 126, 132, 129, 120, 127, 124])
n = len(data)
media = np.mean(data)
desv_std = np.std(data, ddof=1) # desviación estándar muestral (ddof=1)
alpha = 0.05 # nivel de significancia (95% de confianza)
df = n - 1

# Valor crítico de t (dos colas)
t_critico = t.ppf(1 - alpha/2, df)

# Error estándar
error = t_critico * (desv_std / np.sqrt(n))
```

```python
# Intervalo de confianza
lim_inf = media - error
lim_sup = media + error

print(f"Media muestral = {media:.2f} mmHg")
print (f"Desviación estandar = {desv_std:.2f} mmHg")
print (f"Grados de libertad = {df}")
print(f"Valor crítico t = {t_critico:.3f}")
print(f"Intervalo de confianza 95%: [{lim_inf:.2f}, {lim_sup:.2f}] mmHg")

# Graficar
x = np.linspace(media - 3*desv_std, media + 3*desv_std, 200)
y = t.pdf((x - media) / (desv_std / np.sqrt(n)), df)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='navy', label='Distribucion t de la media muestral' )
plt.axvline(lim_inf, color='red', linestyle='--', label='Limite inferior' )
plt.axvline(lim_sup, color='red', linestyle='--', label='Limite superior' )
plt.axvline(media, color='green', linestyle='-', label='Media muestral' )
plt.title("Intervalo de confianza del 95% para la media (t de Student)")
plt.xlabel("Presión arterial (mmHg)")
plt.ylabel("Densidad de probabilidad")
plt.legend ()
plt.grid(True)
plt. show()
```
