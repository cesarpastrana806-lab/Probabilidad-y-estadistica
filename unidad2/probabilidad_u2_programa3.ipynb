```python
import math as mt
from itertools import combinations
from itertools import permutations
from itertools import product
import numpy as np
```

### **Con combinacion **

```python
asistentes = ["Alan", "Barbara", "Cristian", "Diego", "Emilio", "Fernando", "Gabriel", "Hector", "Ines", "Jessica"]
n = len(asistentes)
r = 4

Cn = mt.factorial(n) / (mt.factorial(r) * mt.factorial(n-r))
print("Existen", Cn, " posibles combinaciones sin repeticion del total de ", n ," nombres ", " en grupos de ", r, " en ", r)
print("Aqui las combinaiones")
combinaciones_lista = list(combinations(asistentes, r))

# Imprime las combinaciones
for combinaciones in combinaciones_lista:
    print(combinaciones)
```

### ** Con permutacion**

```python
asistentes = ["Alan", "Barbara", "Cristian", "Diego", "Emilio", "Fernando", "Gabriel", "Hector", "Ines", "Jessica"]
n = len(asistentes)
r = 4

Pr = mt.factorial(n) / mt.factorial(n-r)
print("Existen ", Pr, " posibles permutacions sin repeticion del total de ", n, " nombres", " en grupos de ", r , " en ", r)

print("Aqui las permutaciones ")
permutaciones_nombres = permutations(asistentes, r)

# Imprime las permutaciones
for permutacion in permutaciones_nombres:
  print(permutacion)
```
