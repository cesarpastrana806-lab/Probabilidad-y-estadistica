```python
x=[5,6,7,8,9,10]
p=[0.05,0.10,0.20,0.30,0.25,0.10]
```

```python
print("Suma ded probabilidades:",sum(p))
```

```python
import matplotlib.pyplot as plt

plt.stem(x,p)
plt.title("Distribución de probabilidad")
plt.xlabel("Calificación")
plt.ylabel("Probabilidad")
plt.grid(True)
plt.show()
```
